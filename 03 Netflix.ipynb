###**4. Column Description**

**Column Name |	Description**<br>
**show_id:** Unique identifier assigned to each title<br>
**type:**	Indicates whether the content is a Movie or TV Show<br>
**title:**	Name of the movie or TV show<br>
**director:**	Director(s) of the content; may contain missing values<br>
**cast:**	List of actors featured in the content<br>
**country:**	Country or countries where the content was produced<br>
**date_added:**	Date when the content was added to Netflix<br>
**release_year:**	Year the content was originally released<br>
**rating:**	Content rating indicating target audience (e.g., PG, TV-MA)<br>
**duration:**	Duration in minutes (for movies) or number of seasons (for TV shows)<br>
**listed_in:**	Genre or category classification<br>
**description:**	Brief summary or synopsis of the content<br>

## **Data Analysis**

```python
# Let's Check Which are more produce.

data["type"].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.set_style("whitegrid")
sns.countplot(x='type', data=data, palette="Reds_r")

plt.title("Movies vs TV Shows", fontsize=14, fontweight="bold")
plt.xlabel("Type", fontsize=10, fontweight="bold")
plt.ylabel("Count", fontsize=10, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

sns.despine()
plt.show()
```

##### There are huge difference between movies and tv shows. And we can easily say that Movies are more produce than TV Shows.

```python
mov_release_year=data.loc[data["type"]=="Movie"]
mov_release_year=mov_release_year["release_year"].value_counts().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))

sns.set_style("whitegrid")

# line plot
sns.lineplot(x=mov_release_year.index, y=mov_release_year.values, marker='o', color='red')

plt.title("Movies Release Year", fontsize=14, fontweight="bold")
plt.xlabel("Release Year", fontsize=10, fontweight="bold")
plt.ylabel("Count", fontsize=10, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

sns.despine()
plt.show()
```

```python
show_release_year=data.loc[data["type"]=="TV Show"]
show_release_year=show_release_year["release_year"].value_counts().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))

sns.set_style("whitegrid")

sns.lineplot(x=show_release_year.index, y=show_release_year.values, marker='o', color='red')

plt.title("TV Show Release Year", fontsize=14, fontweight="bold")
plt.xlabel("Release Year", fontsize=10, fontweight="bold")
plt.ylabel("Count", fontsize=10, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

sns.despine()
plt.show()
```

##### Movies graph look good untill 2016, but after hitting 2017 high production there are major downfall and in 2019 it is almost similar to 2015 production which looks very harsh negative impact.

##### TV Shows graph look very possitive for the initial years like between 2010 to 2014 the TV shows market was consistent but after 2014 to until now the TV shows production has increased very harsh and it looks possitive upside.

```python
# Let's Check which country produced more Movies and TV Shows

cont_produce=data.groupby("country")["type"].value_counts().sort_values(ascending=False).head(15)

cont_produce=cont_produce.reset_index(name="count")

plt.figure(figsize=(15,5))

sns.set_style("whitegrid")
sns.barplot(x="country",
            y="count",
            data=cont_produce,
            hue="type",
            palette="Reds_r")

plt.title("Country wise Production", fontsize=14, fontweight="bold")
plt.xlabel("Country", fontsize=10, fontweight="bold")
plt.ylabel("Count", fontsize=10, fontweight="bold")

plt.show()
```

##### Major production of Movies and TV Shows are produce from United States and India as Second Producer.

```python
data["director"].value_counts().head(10)
```

```python
# Let's Check which country produced more Movies and TV Shows

dirc_produce=data.groupby("director")["type"].value_counts().sort_values(ascending=False).head(10)

dirc_produce=dirc_produce.reset_index(name="direct")

plt.figure(figsize=(15,5))

sns.set_style("whitegrid")
colors=sns.color_palette("Reds_r",n_colors=len(dirc_produce))

sns.barplot(x="director",
            y="direct",
            data=dirc_produce,
            palette=colors)

plt.title("Director wise Production", fontsize=14, fontweight="bold")
plt.xlabel("Director", fontsize=10, fontweight="bold")
plt.ylabel("Count", fontsize=10, fontweight="bold")

plt.xticks(rotation=45)
sns.despine()
plt.show()
```

##### The Director Raul Campos, Jan Suter has made more movie untill now and beside there is Marcus Raboy, Jay Karas, Jay Chapman are on same no of range.

```python
# Let's Check which rating are the most given

rating_data=data.loc[data["type"]=="TV Show", "rating"].value_counts().sort_values(ascending=False)
rating_data=rating_data.reset_index(name="Rate")

plt.figure(figsize=(15,5))

sns.set_style("whitegrid")
colors=sns.color_palette("Reds_r",n_colors=len(rating_data))

sns.barplot(x="rating",
            y="Rate",
            data=rating_data,
            palette=colors
            )

plt.title("TV Show Rating", fontsize=14, fontweight="bold")
plt.xlabel("Rating", fontsize=10)
plt.ylabel("Count", fontsize=10)

sns.despine()
plt.show()
```

##### TV-MA has the highest rating in TV Shows and beside that TV-14 has the second highest rating. G has no rating given for TV Show.

```python
# Let's Check the Most Popular Movie Directors From India

mov_dir=data.loc[(data["type"]=="Movie") & (data["country"]=="India"),"director"].value_counts().head(10)
mov_dir

plt.figure(figsize=(15,5))

sns.set_style("whitegrid")

sns.barplot(x=mov_dir.index,
            y=mov_dir.values,
            palette="Reds_r",
            )

plt.title("Most Popular Movie Directors From India", fontsize=14, fontweight="bold")
plt.xlabel("Director", fontsize=10, fontweight="bold")
plt.ylabel("Count", fontsize=10, fontweight="bold")

sns.despine()
plt.show()
```

```python
mov_tv = data.loc[data["country"] == "India", "type"].value_counts()

plt.figure(figsize=(6,6))

colors = ["#ff4d4d", "#990000"]  # light red & dark red

plt.pie(
    mov_tv.values,
    labels=mov_tv.index,
    autopct="%1.1f%%",
    startangle=110,
    colors=colors,
    wedgeprops={"edgecolor": "white", "linewidth": 2}
)

plt.title("Movies vs TV Shows in India", fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
```

##### We can clearly see that India Are Producing more Movies rather than TV Shows
